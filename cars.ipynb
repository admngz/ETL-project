```python
import pandas as pd
```

```python
# load datasets
make_df =  pd.read_csv('Make.csv')
auction_df = pd.read_csv('USA_cars_datasets.csv')
sales_df = pd.read_csv('Car_sales.csv')
msrp_df = pd.read_csv('data.csv')
```

```python
# drop na values
auction_df.dropna(how='all', inplace=True)
```

# Auction

```python
# make_df cleaning steps
auction_df.country.unique()

# drop columns with country=Canada
auction_df2 = auction_df[['Unnamed: 0', 'price', 'brand', 'year', 'color', 'state', 'country', 'mileage']].copy()
```

```python

index = auction_df2[auction_df2['year'] > 2017].index
auction_df2.drop(index, inplace=True)
```

```python
#confirm years are between 1993-2017
auction_df2.year.unique()
```

```python
# remove canada from dataframe
index = auction_df2[auction_df2['country'] == 'canada'].index
auction_df2.drop(index, inplace=True)
```

```python
# noted that country had a space before the usa
auction_df2.country.unique()
```

```python
# remove space
auction_df2.country = auction_df2.country.str.replace(' ', '')
```

```python
auction_df2.country.unique()
```

```python
auction_clean = auction_df2.rename(columns={'Unnamed: 0': 'id'})
```

```python
# export to clean .csv
auction_clean.to_csv(r'auction_clean.csv', index=False)
```

# sales

```python
sales_df
```

```python
# drop NaN values
sales_df = sales_df.dropna()
```

```python
sales_clean = sales_df.rename(columns={'__year_resale_value': 'resale_value'})
```

```python
sales_clean.reset_index(drop=True, inplace=True)
```

```python
sales_clean['Manufacturer'] = sales_clean['Manufacturer'].str.lower()
sales_clean['Model'] = sales_clean['Model'].str.lower()
```

```python
sales_clean
```

```python
# drop last two columns 
sales_clean
```

```python
sales_clean = sales_clean.drop(['Latest_Launch', 'Power_perf_factor', 'Vehicle_type'], axis=1)
```

```python
sales_clean
```

```python
# export to .csv
sales_clean.to_csv('sales_clean.csv', index=True)
```
